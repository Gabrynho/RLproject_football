```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
# Assuming df is your DataFrame with columns ['id', 'x', 'y', 'time']
df = pd.read_csv('episode_4_observations.csv')

# Rimuovo tutte le colonne con valori pari a 0 e -1. o 1.
#print(df.iloc[1])

# Ball coordinates
ball = df[['88', '89']]*100
ball.columns = ['x', 'y']
ball['x']=ball['x']+100
ball['y']=ball['y']+42
#print(ball)

cr7 = df[['2', '3']]*100
cr7.columns = ['x', 'y']
cr7['x']=cr7['x']+100
cr7['y']=cr7['y']+42

gk = df[['0', '1']]*100
gk.columns = ['x', 'y']
gk['x']=gk['x']+100
gk['y']=gk['y']+42

gk_opponent = df[['8', '9']]*100
gk_opponent.columns = ['x', 'y']
gk_opponent['x']=gk_opponent['x']+100
gk_opponent['y']=gk_opponent['y']+42

print(gk_opponent)
```

```python
# Create the pitch
pitch = Pitch(pitch_type='custom',  # example plotting a tracab pitch
              pitch_length=200, pitch_width=84,
              axis=True, label=True, half=False)  # showing axis labels is optional
fig, ax = pitch.draw()
ax.scatter(ball['x'], ball['y'], color='blue', edgecolors='black', zorder=12)
ax.scatter(cr7['x'], cr7['y'], color='red', edgecolors='black', zorder=12)

plt.show()
```
